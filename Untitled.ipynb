```python
import mltools as ml
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import mltools.dtree_z as dt
```

```python
data = np.genfromtxt('data/curve80.txt')
X = data[:, :-1]
Y = data[:,-1:]
```

```python
learner = dt.treeRegress()
```

```python
learner.train(X, Y, 2)
```

```python
lines = plt.plot(X, Y, 'r.')
```

```python
xs = np.linspace(0,10,200)[:,np.newaxis]
ys = learner.predict(xs)
```

```python
print(ys)
```

```python
lines = plt.plot(xs,ys,'k-',X,Y,'r.', linewidth=3,markersize=12)  
```

```python
learner.print_tree()
```

```python
learner.train(X, Y, 4)
ys = learner.predict(xs)
```

```python
lines = plt.plot(xs,ys,'k-',X,Y,'r.', linewidth=3,markersize=12)  
```
